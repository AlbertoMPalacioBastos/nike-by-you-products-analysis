```python
df0['By You']= df0['PRODUCT_NAME'].str.contains('By You')
```

```python
fig, ax = plt.subplots(1, 2, figsize = (12,6))

# Create boxplot 

sns.boxplot(data=df0[df0['PRODUCT_TYPE']=='FOOTWEAR'], 
            y=df0['PRODUCT_NAME'].str.contains('By You'), 
            x='PRICE_RETAIL', 
            hue=df0['By You'],
            orient="h",
            ax=ax[0],
            palette="magma")
ax[0].invert_yaxis()
ax[0].set_xlabel('Retail price')
ax[0].set_ylabel('"By You" line product')
ax[0].set_title('Nike footwear retail price boxplots', fontsize='12')
# plot line showing the mean of 'PRICE_RETAIL' of all products
mean_price = df0[df0['PRODUCT_TYPE']=='FOOTWEAR']['PRICE_RETAIL'].mean()
ax[0].axvline(x=mean_price, color='r', label='averaga price of all products', ls='--')
ax[0].text(25, 0.5, "Average retail price of all footwear",  fontsize='8')

# Create histogram 
sns.histplot(data=df0[df0['PRODUCT_TYPE']=='FOOTWEAR'], x='PRICE_RETAIL', 
             hue=df0['By You'], 
             binwidth=15,
             multiple='dodge', 
             ax=ax[1],
             shrink=0.85,
             palette="magma")
ax[1].set_xlabel('Retail price')
ax[1].set_title('Nike footwear retail price histogram', fontsize='12')

# Display the plots
plt.show()
```
